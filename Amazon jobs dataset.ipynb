```python
# finding number of openings in Bangalore and Seattle
import csv
with open('Amazon_jobs_dataset.csv', encoding='ISO-8859-1') as file_obj:
    file_data=csv.DictReader(file_obj, skipinitialspace=True)
    
    jobs={'Bangalore':0, 'Seattle': 0}
    for row in file_data:
        key1='Bangalore'
        key2='Seattle'
        value=1

        if key1 in row['location']:
            jobs[key1]= jobs[key1]+value
        if key2 in row['location']:
            jobs[key2]= jobs[key2]+value
            
                
    for i in jobs:
        print(jobs[i], end=" ")
        
```
